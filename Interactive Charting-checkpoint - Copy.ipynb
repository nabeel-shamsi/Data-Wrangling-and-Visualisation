```python
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool
from bokeh.io import push_notebook
from bokeh.layouts import column

# Enable Bokeh in notebook
output_notebook()
```

```python
import warnings
warnings.filterwarnings('ignore')

warnings.simplefilter(action='ignore', category=FutureWarning)
df = pd.read_csv(r"C:\Users\nblsh\Downloads\time_series_data.csv")

df['Date'] = pd.to_datetime(df['Date'])  # adjust the column name if needed

df.head()
```

```python
p = figure(x_axis_type = 'datetime',
          title = "Time vs Value Interactive Chart",
          plot_width = 800, plot_height = 400,
          tools = "pan, wheel_zoom, box_zoom, reset, save")

p.line(df['Date'], df['Value'], color = 'blue', line_width = 2, legend_label ='Value')

p.add_tools(HoverTool(tooltips =[('Date', '@x{%F %T}'), ('Value','@y')], formatters = {'@x' : 'datetime'},
                     mode ='vline'
                     ))

p.xaxis.axis_label = 'Date'
p.yaxis.axis_label = 'Value'
p.legend.location = "top_left"

show(p)
```
