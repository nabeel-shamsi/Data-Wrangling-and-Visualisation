```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv(r"C:\Users\nblsh\Downloads\metrics.csv")
```

```python
fig = go.Figure()

fig.add_trace(go.Bar(x=df["Month"], y=df["Revenue"], name="Revenue"))

fig.add_trace(go.Bar(x=df["Month"], y= df["Profit"], name="Profit"))

fig.add_trace(go.Scatter(x=df["Month"], y=df["Revenue"], name="Revenue", mode="lines+markers"))

fig.add_trace(go.Scatter(x=df["Month"], y=df["Profit"], name="Profit", mode="lines+markers"))

fig.update_layout(updatemenus=[dict(type="buttons",
                                   direction="right",
                                   buttons=[dict(label="Bar Chart",
                                                method="update",
                                                args=[{"visible": [True, True, False, False]},
                                                     {"title": "Metrics-BarChart"}]
                                                ),
                                           
                                           dict(label="Line Chart",
                                               method="update",
                                                args=[{"visible": [False, False, True, True]},
                                                     {"title": "Metrics-LineChart"}]
                                               )],
                                   showactive=True,
                                    x=0.5,
                                    y=1.15,
                                    xanchor="center",
                                    yanchor="top"
                                   )],
                 title="Metrics - Bar Chart",
                 xaxis_title="Month",
                  yaxis_title="Value",
                  template = "plotly_white"
                 )
fig.show()
```
