```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\nblsh\Downloads\top_products.csv", encoding ='latin1')
```

```python
df.head()
```

```python
df = df.sort_values("Revenue", ascending = True)
```

```python
plt.figure(figsize = (10,6))
plt.hlines(y = df["Product Name"],
          xmin = 0,
          xmax = df["Revenue"],
          color = 'skyblue')
plt.plot(df["Revenue"], df["Product Name"], "o", color="blue")
plt.xlabel("Revenue")
plt.title("Top-Performing Products by Revenue")
plt.grid(axis='x', linestyle ='--', alpha = 0.7)
plt.tight_layout()
plt.show()
```
