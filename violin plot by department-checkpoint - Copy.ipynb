```python
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display
```

```python

df = pd.read_csv(r"C:\Users\nblsh\Downloads\employee_salaries.csv")
```

```python
df.head()
```

```python
job_roles = df['Job Role'].dropna().unique().tolist()
dropdown = widgets.Dropdown(options = job_roles,
                           description = 'Job Role:',
                           value = job_roles[0]
                           )
```

```python
def plot_violin(selected_role):
    filtered_df = df[df['Job Role'] == selected_role]
    fig = px.violin(filtered_df,
                   x = 'Experience Level',
                   y = 'Salary',
                   box = True,
                   points = 'all',
                   title = f'Salary Distribution for {selected_role} by Experience Level',
                   color = 'Experience Level'
                   )
    fig.show()
    
widgets.interact(plot_violin, selected_role = dropdown)
```
